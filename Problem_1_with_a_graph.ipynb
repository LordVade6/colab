```python
#import the libraries numpy and matplotn
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=100000
x=np.random.uniform(0,1,n)
y=np.random.uniform(0,1,n)
r=np.sqrt(x**2+y**2)
print("r values",r)
inside=sum(r<1)
outside=sum(r>1)
p=inside/n
print("inside =",inside)
print("outside =",outside)
print("probability of inside =",p)
```

```python
pi=4*inside/n
print("estimate for pi is",pi)
```

```python
plt.figure(figsize=(5,5))
plt.xlim(0,1)
plt.ylim(0,1)
plt.xlabel("x")
plt.ylabel("y")
pointstoplot=1000
plt.scatter(x[:pointstoplot],y[:pointstoplot])
plt.plot(np.linspace(0,1),np.sqrt(1-np.linspace(0,1)**2),color="r")
plt.title(r'$y=\sqrt{1-x^2}$');
```
