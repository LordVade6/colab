```python
#import library as numpy
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=100000
meeting = 0 

for k in range(n):
  l=30*np.random.rand()
  h=30*np.random.rand()
  if h-5<l<h+7:
    meeting+=1
print("the probability of Luna and Harry meeting is",meeting/n)
```

```python
n=100000
x=np.random.uniform(0,30,n)
y=np.random.uniform(0,30,n)
```

```python
plt.figure(figsize=(5,5))
plt.xlim(0,30)
plt.ylim(0,30)
plt.xlabel("H")
plt.ylabel("L")
pointstoplot=1000
plt.scatter(x[:pointstoplot],y[:pointstoplot])
plt.plot(np.linspace(5,30),np.linspace(0,25),color='g',label=r'$y=x-5$')
plt.plot(np.linspace(0,30),np.linspace(0,30),color='r',label=r'$y=x$')
plt.plot(np.linspace(0,23),np.linspace(7,30),color='b',label=r'y=x+7')
plt.legend(bbox_to_anchor=(1,0.2))
plt.title("Probability of L and H meeting")
```
